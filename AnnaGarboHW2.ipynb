Considering the SAR and EAR as Species-Area-Relationship and Endemic-Area-Relationship, the functions are reported considering log series Taylor expansion as in the notes. With different exponents of b (where b is the new born rate) the functions has been plotted in order to observe the difference.

P(n), SAR(a) and EAR(a) are defined as:
$$
P(n)=\frac{1}{log(b)}\frac{(1-b)^{n}}{n}
$$
$$
SAR(a)=S(1-\sum_{n=1}^{+\infty}(1-a)^{n}P(n))
$$
$$
EAR(a)=S\sum_{n=1}^{+\infty}a^{n}P(n)
$$

Using log series expansion they became:
$$
SAR(a)=S(1-\frac{log(1-(1-a)(1-b))}{log(b)})
$$
$$
EAR(a)=\frac{S}{log(b)}log(1-a(1-b))
$$

A comparison has been computed considering power-law version of SAR, defined as:
$$
SAR(a)=Sa^{b}
$$

```python
def sar_sum(a,b,s):
    return s*(1-(np.log(1-(1-b)*(1-a))/np.log(b)))
```

```python
def sar_pow(a,b,s):
    return (s)*a**b 
```

```python
def ear_sum(a,b,s):
    return s/np.log(b)*np.log(1-a*(1-b))
```

```python
B=25/100
S=1000
Z=0.25
A=np.arange(0,1,0.01)
A0=A/25
```

```python
plt.plot(A,sar_sum(A,B,S),label='SAR sum B=0.25')
plt.plot(A,sar_pow(A,B,S),label='SAR B=0.25')
plt.plot(A,ear_sum(A,B,S),label='EAR sum B=0.25')

plt.xlabel('Fraction AREA')
plt.ylabel('Number of Species')
plt.grid(True)
plt.legend()
plt.show()
```

As mentioned before, value of B has been changed to observe the difference, indeed we can see that increasing new born rate three curves come every time closer.

```python
B=np.arange(0.1,1,1/9)
B_re=B.reshape(3,3)

fig,axs=plt.subplots(3,3,figsize=(20,15))
for i in range(3):
    for j in range(3):
        b_value = B_re[i, j]  # Estrai il valore di B per questa iterazione
        axs[i, j].plot(A, sar_sum(A, b_value, S), label=f'SAR sum (B={b_value:.2f})')
        axs[i, j].plot(A, sar_pow(A, b_value, S), label=f'SAR z={b_value:.2f}')
        axs[i, j].plot(A, ear_sum(A, b_value, S), label=f'EAR sum (B={b_value:.2f})')

        axs[i, j].set_xlabel('Fraction AREA')
        axs[i, j].set_ylabel('Number of Species')
        axs[i, j].grid(True)
        axs[i, j].legend()

plt.show()
```
